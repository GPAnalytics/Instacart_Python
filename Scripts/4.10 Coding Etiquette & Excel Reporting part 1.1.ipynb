# 03. Creating Regional Segmentation of the Data and determine differences in spending habits between the different U.S. regions.

```python
# Defining Regions
Northeast = ['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']
Midwest = ['Wisconsin', 'Michigan','Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']
South = ['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']
West = ['Idaho', 'Montana', 'Wyoming', 'Nevada', 'Utah', 'Colorado', 'Arizona', 'New Mexico', 'Alaska', 'Washington', 'Oregon', 'California', 'Hawaii']
```

# 04. Creating an exclusion flag for low-activity customers (customers with less than 5 orders) and excluding them from the data

```python
# To create activity flag, we are using the following criteria:
# if the customer has less than 5 orders, then the customer should be labeled as “low active customer.”
# If the customer has 5 or more orders, then the customer should be labeled as “high active customer.”
```

```python
# Creating flag for customers with 5 or more orders.
ords_prods_cust_merged.loc[ords_prods_cust_merged['max_order'] >= 5, 'customer_activity'] = 'High active'
```

```python
# Creating flag for customers with less than 5 orders
ords_prods_cust_merged.loc[ords_prods_cust_merged['max_order'] <5, 'customer_activity'] = 'Low active'
```

```python
#checking the shape
ords_prods_cust_merged.shape
```

```python
#checking the values in customer activity 
ords_prods_cust_merged['customer_activity'].value_counts(dropna = False)
```

```python
# in order to count each user only once
ords_prods_cust_merged.groupby(['customer_activity']).user_id.nunique()
```

This means that we will remove 43578 users from further analysis

```python
#defining the dataframe with low active customers/creating subset
low_active_customers = ords_prods_cust_merged[ords_prods_cust_merged['customer_activity'] == 'Low active']
```
