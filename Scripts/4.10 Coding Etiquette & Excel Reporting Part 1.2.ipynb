# 5d. Customer Profiling based on dependants status

```python
# Checking the dependents column.
df_active_cust_dpt['number_of_dependants'].describe()
```

Criteria for family status flag:
if divorced/widowed', 'living with parents and siblings', 'single' and no dependants, then 'Single no Dependants',
if divorced/widowed', 'living with parents and siblings', 'single' and numberof dependants >=1, then 'Single with  Dependants' 
if married and and no dependants, then 'Married with no Dependants'(in the data set for every married person number of dependants is set to at least 1),
if married and no_of_dependants >1, then 'Married with  Dependants.


```python
# creating a flag based on the number of dependants
df_active_cust_dpt.loc[(df_active_cust_dpt['marital_status'].isin(['divorced/widowed', 'living with parents and siblings', 'single'])) & (df_active_cust_dpt['number_of_dependants']==0), 'Family_Status_Flag'] = 'Single no Dependants'
```

```python
df_active_cust_dpt.loc[(df_active_cust_dpt['marital_status'].isin(['divorced/widowed', 'living with parents and siblings', 'single'])) & (df_active_cust_dpt['number_of_dependants']>=1), 'Family_Status_Flag'] = 'Single with Dependants'
```

```python
df_active_cust_dpt.loc[(df_active_cust_dpt['marital_status'] == 'married') & (df_active_cust_dpt['number_of_dependants'] == 1), 'Family_Status_Flag'] = 'Married with no Dependants'
```

```python
df_active_cust_dpt.loc[(df_active_cust_dpt['marital_status'] == 'married') & (df_active_cust_dpt['number_of_dependants'] >1), 'Family_Status_Flag'] = 'Married with  Dependants'
```

```python
# checking the frequency of the flag

df_active_cust_dpt['Family_Status_Flag'].value_counts(dropna=False)
```

```python
# in order to count each user only once
df_active_cust_dpt.groupby(['Family_Status_Flag']).user_id.nunique()
```

```python
#checking the dataframe
df_active_cust_dpt.head()
```

```python
# checking the 'Family_Status_Flag' with the 'order_hour_of_day'
family_status_hour_of_day=pd.crosstab(df_active_cust_dpt['order_hour_of_day'],df_active_cust_dpt['Family_Status_Flag'],dropna=False)
```
