# 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Aggregating Data with transform()
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# checking the output
ords_prods_merge.head(20)
```

```python
# To create a loyalty flag, we are using the following criteria:

# 1. If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
# 2. If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
# 3. If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# checking the columns 'user_id', 'loyalty_flag', 'order_number' and 'max_order' 
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number', 'max_order' ]].head(20)
```

# 6.  Create a spending flag for each user based on the average price across all their orders 

```python
# Aggregating Data with transform() and create new column 'user_avg_price'
ords_prods_merge['user_avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# checking the output
ords_prods_merge.head(15)
```

```python
# To create a spending  flag, we are using the following criteria:
# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
```

```python
ords_prods_merge.loc[ords_prods_merge['user_avg_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['user_avg_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# check the frequency
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# checking the columns 'user_id', 'spending_flag', 'user_avg_price'
ords_prods_merge[['user_id', 'spending_flag', 'user_avg_price']].head(50)
```

# 7. Determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column

```python
 # Aggregating Data with transform() and create new column 'median_prior_orders'
ords_prods_merge['median_prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_last-order'].transform(np.median)
```

```python
# checking the output
ords_prods_merge.head(15)
```

```python
# To create an order frequency flag, we are using the following criteria:
# if the median of “days_since_last_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
# If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_prior_orders'] > 10) & (ords_prods_merge['median_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# check the frequency
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Use the 'isna()' function to filter rows with NaN values in the 'order_frequency_flag' column
nan_rows = ords_prods_merge[ords_prods_merge['order_frequency_flag'].isna()]
```

```python
nan_rows
```

```python
# The NaN values are all related to one user and his first order, so this user can be categorized as a "Non-frequent customer.
```

```python
# Changing the null values in "order_frequency_flag" column to "Non-frequent customer"
ords_prods_merge['order_frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
# check the frequency
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
