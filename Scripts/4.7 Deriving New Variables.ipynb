# User-defined function

```python
# define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#  check the values in the new column
df['price_range'].value_counts(dropna=False)
```

```python
# check the max value
df['prices'].max()
```

# If-Statements with the loc() Function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#  check the values in the new column
df['price_range_loc'].value_counts(dropna=False)
```

```python
# repeating the process on the entire dataframe
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

# If-Statements with For-Loops

```python
# printing the frequency of the “orders_day_of_week” column
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# create the blank list and then the loop
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# printing 'result'
result
```

```python
# create a new column within  df_ords_prods_merge dataframe and set it equal to 'result'
df_ords_prods_merge['busiest_day']=result
```

```python
# print the frequency of the new column
df_ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

# 2. Creating Busiest days” column. This label should correspond with the two busiest days of the week as opposed to the single busiest day

```python
# create the blank list and then the loop
result1 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result1.append("Busiest days")
  elif value == 4 or value == 3:
    result1.append("Slowest days")
  else:
    result1.append("Regularly busy")
```

```python
# create a new column within  df_ords_prods_merge dataframe and set it equal to 'result1'
df_ords_prods_merge['busiest_days']=result1
```

```python
# checking the result
df_ords_prods_merge.head()
```

#  4. Identify the busiest hours of the day

```python
# printing the frequency of the “order_hour_of_day” column
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

We can group the 24 hours of the day as follows: Most orders = 10, 11, 14, 15, 13, 12, 16, 9; Average orders = 17, 8, 18, 19, 20, 7, 21, 22; Fewest orders = 23, 6, 0, 1, 5, 2, 4, 3

```python
# create the blank list and  the loop

result2 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result2.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result2.append("Fewest orders")
  else:
    result2.append("Average orders")

```

```python
# create a new column within  df_ords_prods_merge dataframe and set it equal to 'result2'
df_ords_prods_merge['busiest_period_of_day']=result2
```
